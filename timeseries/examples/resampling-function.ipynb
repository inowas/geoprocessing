```python
import json
import pandas as pd

def is_json(j):
  try:
    json.loads(j)
  except ValueError as e:
    return False
  
  return True

def parse_input(d):
    if not is_json(d):
        return False
    
    df = pd.read_json(d)
    if 'timeStamp' in df.columns:
        return df.set_index('timeStamp')
    
    return False

def resample(df, rule='1D', interpolation_method='linear'):
    """
    :param df: Pandas dataframe with time-based index
    :param rule: pandas resample rule
    :param interpolation_method: 
    """
    
    if not isinstance(df, pd.DataFrame):
        return False
    
    return df.resample(rule).interpolate(method=interpolation_method)
```
