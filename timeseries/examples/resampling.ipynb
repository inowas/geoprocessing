## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
time_series['value'].plot(linewidth=0.5)
```

## Resampling

```python
time_series_1 = time_series.resample("6H").asfreq()
time_series_1['value'].plot(linewidth=0.5)
time_series_1.head()
```

```python
time_series_2 = time_series.resample("6H").ohlc()
time_series_2['value'].plot(linewidth=0.5)
time_series_2.head()
```

```python
time_series = time_series.resample("H").ffill()
time_series['value'].plot(linewidth=0.5)
```
